```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
# these features combaine two tables
Final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
Final_df.head(3)
```

```python
# To check data types
Final_df.dtypes
```

```python
# Explore datasets
# How many country in these datasets to see from these function
Final_df.Country.value_counts()
```

```python
# create a country_name variable
country_name=Final_df.Country.value_counts().index
```

```python
# array form all these country
country_val=Final_df.Country.value_counts().values
```

```python
# pie chart - Top 3 countries that uses zomato
plt.pie(country_val[:3],labels=country_name[:3])
```

```python
# percentages of Top 3 countries that uses zomato in percentages
plt.pie(country_val[:3],labels=country_name[:3], autopct='%1.2f%%')
```
